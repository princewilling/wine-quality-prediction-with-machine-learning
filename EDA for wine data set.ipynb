```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_wine = pd.read_csv("dataset/wine.data", header=None)
```

```python
df_wine.columns = ['class','Alcohol','Malic acid','Ash','Alcalinity of ash',  'Magnesium','Total phenols',
                   'Flavanoids','Nonflavanoid phenols',
                   'Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline']
```

```python
df_wine.tail()
```

```python
df_wine.info()
```

```python
df_wine.describe(include='all')
```

```python
df_wine.isnull().sum()
```

```python
plt.subplot(221)

df_wine['class'].value_counts().plot(kind='bar', title='Class Ratio', figsize=(16,9))

plt.xticks(rotation=0)
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(df_wine, hue="class", size=5).map(plt.scatter, "Ash" ,"Alcalinity of ash").add_legend()
plt.show()
```

```python
sns.set_style("whitegrid");
sns.pairplot(df_wine, hue="class", size=3);
plt.show()
```

```python
plt.figure(figsize=[20,10])
corr = df_wine.corr()
sns.heatmap(corr, annot=True, square=True)
plt.yticks(rotation=0)
plt.show()
```
